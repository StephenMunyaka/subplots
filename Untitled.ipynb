```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style
%matplotlib inline
```

```python
wind_data = [34,56,77,43,10,33,79,12,15]
rain_data = [20,50,72,34,67,65,30,56,44]
time_hrs = [2,3,4,5,6,7,8,9,10]
```

```python
plt.plot(time_hrs,wind_data,'g',label='wind',linestyle='--',linewidth=1)
plt.plot(time_hrs,rain_data,'b',label='rain',linestyle='--',linewidth=1.5)
plt.title('weather traffic')
plt.xlabel('Hrs')
plt.ylabel('weather')
plt.legend()
plt.show()
```
