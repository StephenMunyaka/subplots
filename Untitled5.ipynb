```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style
%matplotlib inline
web_monday = [10,4,8,19,1]
web_tuesday = [5,9,11,6,2]
time_hrs = [1,2,3,4,5]
style.use('ggplot')
plt.figure(figsize=[11,6])
plt.subplots_adjust(hspace=.25)
plt.subplot(1,2,1)
plt.plot(time_hrs,web_monday,'y',label='monday',linestyle='--',linewidth=2)
plt.title('MONDAY')
plt.xlabel('Hrs')
plt.ylabel('Number of people')
plt.subplot(1,2,2)
plt.plot(time_hrs,web_tuesday,'b',label='tuesday',linestyle='--',linewidth=1)
plt.title('TUESDAY')
plt.xlabel('Hrs')
plt.ylabel('Number of people')
plt.legend()
plt.show()
```
