```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style
%matplotlin inline
```

```python
web_monday=[23,56,45,33,45,10,5]
web_tuesday=[40,5,23,60,33,15,10]
time_hrs=[1,2,3,4,5,6,7]
```

```python
style.use('ggplot')
plt.plot(time_hrs,web_monday,'r',label='monday',linestyle='--',linewidth=1)
plt.plot(time_hrs,web_tuesday,'r',label='tuesday',linestyle='--',linewidth=2)
plt.title('Attendance per hour')
plt.xlabel('Number of users')
plt.ylabel('Hrs')
plt.legend()
plt.show
```
